# Average Values of reviews

```python
df['Reviews'].dtype
```

```python
df['Reviews'].astype('float')
```

#### some rows have  3m as their reviews, we need to remove them 

```python
df[df['Reviews'] == '3.0M']
```

```python
df['Reviews'] = df['Reviews'].replace('3.0M', 3.0)
```

```python
df['Reviews'] =  df['Reviews'].astype('float')
```

```python
df['Reviews'].dtype
```

```python
df['Reviews'].mean()
```
